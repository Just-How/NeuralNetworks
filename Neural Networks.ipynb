```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

#1: Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#2: Split data into training and testing set, with test size of 0.30
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)

#3: Create a neural network for regression using sklearn Use a hidden layers of size 130 -> 64 -> 32 -> 16, and activation function 'relu'
mlp_regressor = MLPRegressor(hidden_layer_sizes=(130, 64, 32, 16), activation='relu', random_state=42)

# Fit the model to the training data
mlp_regressor.fit(X_train, y_train)

#4: Run the test data through the model to obtain predictions.
predictions = mlp_regressor.predict(X_test)

#5: Print the R2 of your model
r2 = r2_score(y_test, predictions)
print("R2 Score:", r2)
```

```python
print("How did your model perform?\n  This R2 score indicates that the model explains approximately 79.1% of the variance in the house values. Model performs well in predicting house values based on features.")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'hidden_layer_sizes': [(130, 64, 32, 16), (100, 50, 25), (150, 75, 37, 18)],'activation': ['relu', 'tanh', 'logistic']}

mlp_regressor = MLPRegressor(random_state=42)

grid_search = GridSearchCV(mlp_regressor, param_grid, cv=3, scoring='r2')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

print("GridsearchCV Parameters:", best_params)

best_predictions = best_model.predict(X_test)
best_r2 = r2_score(y_test, best_predictions)

print("GridsearchCV R2 Score:", best_r2)
```
